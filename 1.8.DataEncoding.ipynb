# <b>Data Encoding
    
  Data Encoding is necessary to convert categorical data into continuous data. Because ML models work only continuous datasets.   
  Data encoding is a data transformation which changes the structure of data. So it not so same as Standardization. \
    Dummy Variable Trap is another issue in multi-class categorical variable that will be explained below
    
    
  1) Label Encoding / 1-0 Transformation \
  2) 1 and Others 0 Transformation / np.where() \
  3) One Hot Encoding / Multi-class Transformation  

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df= sns.load_dataset("tips")
df.head()
```

## <b>1)  Label Encoding / 1 - 0 Transformation 
    
    This is the most useful method.
    It converts categorical values like male-female, noon-evening into 1 and 0
    Warning: We can think to give it 1 which value is the more important.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LabelEncoder().fit_transform(df["sex"])
```

```python
df["sex_new"] = LabelEncoder().fit_transform(df["sex"])
df
```

## <b>2) 1 and Others 0 Transformation / np.where() 

```python
df.head()
```

```python
df["day_new"] = np.where(df["day"].str.contains("Sun"), 1, 0)
df
```

## <b>3) One Hot Encoding /Multi-Class Transformation 

```python
df.head()
```

```python
LabelEncoder().fit_transform(df["day"])
```

Warning: Each number above represent each day in dataset. This kind of numbering will cause ML model to understand ordinal relation between days. This issue calls ***Dummy Variable Trap***. 
Then we should use ***One Hot Encoding***

- ### One Hot Encoding / pd.get_dummies()

```python
df= sns.load_dataset("tips")
df_one_hot = pd.get_dummies(df, columns=["day"],prefix=["day"])
df_one_hot.head()
```
