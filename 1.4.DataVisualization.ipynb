### How to reorder Wrong Ordinal Data

* Let's teach to Python that there is an ordinary in cut variable

```python
from pandas.api.types import CategoricalDtype
```

```python
cut_categories =["Fair", "Good", "Very Good", "Premium", "Ideal"]
```

```python
df["cut"]=df["cut"].astype(CategoricalDtype(categories=cut_categories, ordered = True))
```

```python
df["cut"].head(1)
```

```python
df["cut"].value_counts().plot.bar().set_title("Cut Variables Frequencies")
```

```python
sns.barplot(x="cut", y=df["cut"].index, data=df)
```

### Barplot Transposition: 
- It provides additional detailed information

```python
df.head()
```

- sns.catplot()
- sns.barplot()

```python
sns.catplot(x="cut", y="price", data=df)
```

```python
sns.barplot(x="cut",y="price", hue="color",data=df)
```

#### The price looks around 17K in the first visualization, but 6K in the second one. Why?
- Seaborn represent sum in Y-axis in first visualization and mean in Y-axis in Second one.  

- Black lines on bars represent standard deviations / standard errors.

```python
df.groupby(["cut","color"])["price"].mean()
```

### Histogram and Intensity Transposition

Graphs tell something while hiding something. We need to find hiding details using Transposition

```python
sns.FacetGrid(df, 
              hue="cut",
              height= 5,
              xlim=(0,10000)).map(sns.kdeplot, 
                                  "price", shade=True).add_legend()
```

```python
sns.catplot(x="cut", y="price", hue="color", kind="point", data=df); #hue brings 3rd variable into graph
```

We can undestand there is a cut-price correlation. Cut independent variable is affecting Price dependent variable.

```python
# This is important to remember there might be effect of changing conditions in time.

df["passengers"].describe()
```

### We should prepare data before applying heatmap.

We will utilize Pivot

- df.pivot()

```python
df.pivot("month", "year", "passengers")
```

```python
df.pivot_table("passengers", index="month", columns="year")
```

```python
df_pivot =df.pivot("month", "year", "passengers")
```

```python
sns.heatmap(df_pivot);

# It help us see pattern!
```

We can understand results below from the heatmap graph:
- Travelling with plane increased over the years.
- Number of passangers increase in June, July and August 

```python
sns.heatmap(df_pivot, annot=True, fmt="d")
```

```python
sns.heatmap(df_pivot, annot=True, fmt="d", linewidths=.5);
```

```python
sns.heatmap(df_pivot, annot=True, fmt="d", linewidths=.5, cbar=False);
```
